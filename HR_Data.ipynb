which jon title had a highest avergae salary

```python
salary=df.groupby("Job_Title")["Salary_INR"].mean()/1000
```

```python
salary.plot(kind="bar")
plt.xlabel("Job Title")
plt.ylabel("Average Salary")
plt.show()
```

what is the avergae salary in different department based om job title

```python
dep_job=df.groupby(["Department","Job_Title"])["Salary_INR"].mean()/1000
```

```python
dep_job.plot(kind='barh',figsize=(15,10), color='skyblue')
plt.title("Average Salary by Department and Job Title")
plt.xlabel("Average Salary")
plt.ylabel("Department and Job Title")
plt.show()
```

how many employee resigned and terminated in each department

```python
df.Status.unique()
```

```python
reg=df[df["Status"]=="Resigned"].groupby("Department")["Status"].count()
reg
```

```python
ter=df[df["Status"]=="Terminated"].groupby("Department")["Status"].count()
ter
```

```python
reg.plot(x=reg.index,y=reg.values,kind="bar",color="black",legend=True,label="regined")
ter.plot(x=ter.index,y=ter.values,kind="bar",color="orange",legend=True,label="terminated")
plt.show()
```
